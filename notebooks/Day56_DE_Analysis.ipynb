#### Subsubtask 1.3.1: Implement a ratio intensity plot (using matplotlib)

```python
def ratio_intensity_plot(x, y):
    """
    Computes a Ratio-Intensity-Plot for two given
    phenotypes from the data loaded in the previous task.
    """
    
    # convert to array to perform calculations
    phen1 = np.array(x)
    phen2 = np.array(y)
    
    # compute R values
    R = np.log10(phen1/phen2)
    
    # compute I values
    I = (np.log10(phen1) + np.log10(phen2)) / 2
    
    # regression curve
    coefs = poly.polyfit(I, R, 4)
    ffit = poly.polyval(I, coefs)
    
    # general settings
    plt.figure(figsize=(10,6))
    plt.xlabel = "Intensity"
    plt.ylabel = 'Ratio'
    plt.title('Ratio-Intensity-Plot')
    
    # line on y=0
    plt.axhline(color='blue')
    
    # produce simple and unstyled plot for testing
    plt.scatter(I,R, color='k', s=1)
    plt.plot(I, ffit, 'r.')
    
    plt.show()
```

#### Subsubtask 1.3.3: Implement a RLE plot (RLE = Relative Log Expression)

```python
def relative_log_expression_plot(data):
    
    dcopy = data
    
    # compute averages
    avg = dcopy/dcopy.mean()
    
    # set plot size
    ax = plt.figure(figsize=(16,12), dpi=80).add_subplot(111)
    ax.set_title('Relative Log Expression Plot')
    plt.xticks(range(len(data)), data.index,
           rotation="vertical", fontsize=10)
    ax.set_xlabel('Gene')
    ax.set_ylabel('Relative Log Expression')
    
    # plot boxplots using seaborn
    sns.boxplot(data=np.log10(avg))
    
    
    plt.show()
```

#### Subsubtask 2.1.2: Make a Q-Q plot comparing the two samples for the just chosen gene and answer the question: Does this tell you anything?

```python
def qqplot(x_, y_, title='QQ Plot'):
    
    # prepare plot
    plt.figure(figsize=(10,6), dpi=100)
    ax = plt.gca()
    ax.set_title(title)
    ax.set_xlabel('Quantile 1st sample')
    ax.set_ylabel('Quantile 2nd sample')
    plt.grid(True, color="lightgrey")
    
    # Plot qq-plot and 45° angle bisector
    qqplot_2samples(x_.values, y_.values, ax=ax, line="45")
    plt.show()
```

```python
# get CD diet values
cd_diet = random_row.filter(regex='CD')

# get HFD diet values
hfd_diet = random_row.filter(regex='HFD')

# produce qq-plot
qqplot(hfd_diet, cd_diet, title='QQ Plot HFD vs CD')
```
